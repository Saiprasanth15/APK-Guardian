```python
import pandas as pd
df=pd.read_csv('weka_data.csv')
df
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
x=df.drop('class',axis=1)
y=df[['class']]
```

```python
from sklearn.svm import SVC
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
sfs_svm = SFS(SVC(),
         k_features=25,
         forward=True,
         floating=False,
         scoring='accuracy',
         cv=0)
#sfs_svm.fit(x, y)
#sfs_svm.k_feature_names_
```

```python
import warnings;
warnings.simplefilter('ignore')
pipe_sfs_svm = make_pipeline(StandardScaler(), sfs_svm)

pipe_sfs_svm.fit(x, y)

print('best combination (ACC: %.3f): %s\n' % (sfs_svm.k_score_, sfs_svm.k_feature_idx_))
print('all subsets:\n', sfs_svm.subsets_)
plot_sfs(sfs_svm.get_metric_dict(), kind='std_err');
```

```python
from sklearn.naive_bayes import GaussianNB
sfs_nb = SFS(GaussianNB(), 
           k_features=100, 
           forward=True, 
           floating=False, 
           scoring='accuracy',
           cv=4,
           n_jobs=-1)
#sfs_nb = sfs_nb.fit(x, y)
#sfs_nb.k_feature_names_
```

```python
sfs_nb = sfs_nb.fit(x, y)
sfs_nb.k_feature_names_
```

```python
pipe_sfs_nb = make_pipeline(StandardScaler(), sfs_nb)

pipe_sfs_nb.fit(x, y)

print('best combination (ACC: %.3f): %s\n' % (sfs_nb.k_score_, sfs_nb.k_feature_idx_))
print('all subsets:\n', sfs_nb.subsets_)
plot_sfs(sfs_nb.get_metric_dict(), kind='std_err');
```
